### Data visualization using Histogram


```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
import matplotlib.pyplot as plt
```

```python
data =  pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
data
```

```python
# @title Histogram of Sepal Length
plt.hist(data['SepalLengthCm'], bins=10, edgecolor='w', alpha=0.5 )
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# @title Histogram of Sepal width
plt.hist(data['SepalWidthCm'], bins=10, edgecolor='w', alpha=0.75 )
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
# @title Histogram of Petal Length
plt.hist(data['PetalLengthCm'], bins=10, edgecolor='w', alpha=0.5 )
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
# @title Histogram of Petal width
plt.hist(data['PetalWidthCm'], bins=10, edgecolor='w', alpha=1 )
plt.xlabel('Petal width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal width')
plt.show()
```

```python
species_count = data['Species'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
plt.show()
```
